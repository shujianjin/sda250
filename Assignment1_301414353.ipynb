```python
def lexDiv(types, tokens):
    diversity = types/tokens
    print("Lexical Devisity=", diversity)
    return diversity
```

```python
import os
import nltk
#calculate tokens and types 

data = [corpus1, corpus2, corpus3]

for file in data:
    
    print("types=", len((set(file))))
    print("tokens=", len(file) )
    sorted(set(file))
    types= len(set(file))
    tokens= len(file)
    lexDiv(types, tokens)
    print("")
```
